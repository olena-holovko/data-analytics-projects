```python
import pandas as pd

# Завантажуємо основний файл з даними
df = pd.read_csv("survey_results_public.csv")

# Завантажуємо файл зі схемою
schema = pd.read_csv("survey_results_schema.csv")

print("Дані завантажено!")
```

```python
# Завдання 1. Загальна кількість респондентів
total = len(df)
print(f"Загальна кількість респондентів: {total}")
```

```python
# Отримуємо список питань з файлу схеми
questions = schema["qname"].tolist()
print(f"Кількість питань у схемі: {len(questions)}")
```

```python
# Завдання 2. Аналіз повноти відповідей респондентів
# Знаходимо перетин питань схеми і колонок датасету
common_columns = list(set(questions) & set(df.columns))
print(f"Спільних колонок: {len(common_columns)}")
```

```python
# Підраховуємо респондентів без пропущених значень
complete = df[common_columns].dropna().shape[0]
print(f"Респондентів без пропущених відповідей: {complete}")
```

```python
# Завдання 3. Статистичний аналіз досвіду респондентів

mean_exp = df["WorkExp"].mean()
median_exp = df["WorkExp"].median()
mode_exp = df["WorkExp"].mode()[0]

print(f"Середнє: {mean_exp:.1f} років")
print(f"Медіана: {median_exp} років")
print(f"Мода: {mode_exp} років")
```

```python
# Досліджуємо колонку RemoteWork перед фільтрацією
# Дивимося унікальні значення колонки RemoteWork
print(df["RemoteWork"].value_counts())
```

```python
# Завдання 4. Аналіз віддаленої роботи
remote_count = df[df["RemoteWork"] == "Remote"].shape[0]
print(f"Кількість віддалених працівників: {remote_count}")
```

```python
# Завдання 5. Визначення популярності Python
python_count = df["LanguageHaveWorkedWith"].str.contains("Python", na=False).sum()
python_percent = python_count / len(df) * 100

print(f"Кількість Python розробників: {python_count}")
print(f"Відсоток: {python_percent:.1f}%")
```

```python
# Завдання 6. Визначення кількості людей, які навчалися онлайн
online_count = df["LearnCode"].str.contains("Online Course", na=False).sum()
online_percent = online_count / len(df) * 100

print(f"Кількість людей, які вчились онлайн: {online_count}")
print(f"Відсоток: {online_percent:.1f}%")
```

```python
# Завдання 7. Крок 1 - фільтруємо Python-розробників
python_df = df[df["LanguageHaveWorkedWith"].str.contains("Python", na=False)]
print(f"Кількість Python-розробників: {len(python_df)}")
```

```python
# Завдання 7. Крок 2 - групування по країнах
compensation_by_country = python_df.groupby("Country")["ConvertedCompYearly"].agg(
    середня="mean",
    медіанна="median"
).round(0).reset_index()

compensation_by_country = compensation_by_country.sort_values(
    "середня", ascending=False
)
print(compensation_by_country)
```

```python
# Завдання 8. Аналіз освіти найбільш оплачуваних спеціалістів
top5 = df.sort_values("ConvertedCompYearly", ascending=False).head(5)
print(top5["EdLevel"])
```

```python
# Дивимося унікальні вікові групи
print(df["Age"].value_counts())
```

```python
# Завдання 9. Популярність Python по вікових групах

# Додаємо колонку True/False - чи використовує людина Python
df["uses_python"] = df["LanguageHaveWorkedWith"].str.contains("Python", na=False)

# Групуємо по віку і рахуємо відсоток
python_by_age = df.groupby("Age")["uses_python"].mean() * 100
python_by_age = python_by_age.round(1).sort_values(ascending=False)

print(python_by_age)
```

```python
# Завдання 10. Індустрії високооплачуваних віддалених працівників

# Крок 1 - знаходимо 75 перцентиль
percentile_75 = df["ConvertedCompYearly"].quantile(0.75)
print(f"75-й перцентиль компенсації: {percentile_75}")

# Крок 2 - фільтруємо високооплачуваних віддалених працівників
top_remote = df[
    (df["ConvertedCompYearly"] > percentile_75) &
    (df["RemoteWork"] == "Remote")
]

# Крок 3 - знаходимо найпопулярніші індустрії
top_industries = top_remote["Industry"].value_counts().head(10)
print(top_industries)
```
